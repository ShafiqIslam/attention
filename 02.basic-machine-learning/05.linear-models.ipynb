##### Linear Regression

**Linear regression** is a statistical model that studies the relationship between a dependent variable (Y) with a given set of independent variables (X). The relationship can be established with the help of fitting a best line.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

Provide the values for independent variable X

```python
X = np.array([[1,1],[1,2],[2,2],[2,3]])
X
```

The value of dependent variable y can be calculated as follows

```python
y = np.dot(X, np.array([1,2])) + 3
y
```

Visualize the data

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(X, columns=["X1", "X2"])
df["y"] = y

sns.lmplot(x="X1", y="y", data=df, ci=None, line_kws={"color": "red"}, scatter_kws={"color": "blue"})
plt.xlabel("X1")
plt.ylabel("y")
plt.title("Linear Regression")
plt.show()
```

Create a linear regression object

```python
regr = LinearRegression(
   fit_intercept = True, copy_X = True, n_jobs = 2
).fit(X, y)
regr
```

Predict with the model 

```python
regr.predict(np.array([[3,5]]))
```

```python
regr.score(X,y)
```

```python
regr.coef_
```

```python
regr.intercept_
```

##### Logistic Regression

```python
from sklearn import linear_model
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y = True)
LRG = linear_model.LogisticRegression(
   random_state = 0,solver = 'liblinear'
).fit(X, y)

LRG.score(X, y)
```

##### Ridge Regression

```python
from sklearn.linear_model import Ridge
import numpy as np
n_samples, n_features = 15, 10
rng = np.random.RandomState(0)
y = rng.randn(n_samples)
X = rng.randn(n_samples, n_features)
rdg = Ridge(alpha = 0.5)
rdg.fit(X, y)
rdg.score(X,y)
```

##### Bayesian Ridge Regression

```python
from sklearn import linear_model
X = [[0, 0], [1, 1], [2, 2], [3, 3]]
Y = [0, 1, 2, 3]
BayReg = linear_model.BayesianRidge()
BayReg.fit(X, Y)
```

```python
BayReg.predict([[1,1]])
```

```python
BayReg.coef_
```

##### LASSO (Least Absolute Shrinkage and Selection Operator)

```python
from sklearn import linear_model
Lreg = linear_model.Lasso(alpha = 0.5)
Lreg.fit([[0,0], [1, 1], [2, 2]], [0, 1, 2])
```

```python
Lreg.predict([[0,1]])
```

```python
print("Weight vector:", Lreg.coef_)
print("Intercept:", Lreg.intercept_)
print("Number of iterations:", Lreg.n_iter_)
```

##### Multi-task LASSO

```python
from sklearn import linear_model
MTLReg = linear_model.MultiTaskLasso(alpha=0.5)
MTLReg.fit([[0,0], [1, 1], [2, 2]], [[0, 0],[1,1],[2,2]])
```

```python
MTLReg.predict([[0,1]])
```

```python
print("Weight vector:", MTLReg.coef_)
print("Intercept:", MTLReg.intercept_)
print("Number of iterations:", MTLReg.n_iter_)
```

##### Elastic-Net

```python
from sklearn import linear_model
ENreg = linear_model.ElasticNet(alpha = 0.5,random_state = 0)
ENreg.fit([[0,0], [1, 1], [2, 2]], [0, 1, 2])
```

```python
ENreg.predict([[0,1]])
```

```python
print("Weight vector:", ENreg.coef_)
print("Intercept:", ENreg.intercept_)
print("Number of iterations:", ENreg.n_iter_)
```

##### MultiTaskElasticNet
