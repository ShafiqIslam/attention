All machine learning algorithms in Scikit-Learn are implemented via Estimator API. The object that learns from the data (fitting the data) is an estimator. It can be used with any of the algorithms like classification, regression, clustering or even with a transformer, that extracts useful features from raw data.

**Example of using Estimator API with Supervised Learning**

Create sample data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(35)
x = 10*rng.rand(40)
y = 2*x-1+rng.randn(40)
plt.scatter(x,y)
```

Choose model

```python
from sklearn.linear_model import LinearRegression
```

Choose model hyperparameters

```python
model = LinearRegression(fit_intercept = True)
model
```

Reshape the feature matrix `X` to make it a matrix of size `[n_samples, n_features]`

```python
X = x[:, np.newaxis]
X.shape
```

Apply the model to data

```python
model.fit(X, y)
```

Show the slope of the simple linear fit of the data

```python
model.coef_
```

Show the intercept of the simple linear fit

```python
model.intercept_
```

Apply the model to new data

```python
xfit = np.linspace(-1, 11)
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
plt.scatter(x, y)
plt.plot(xfit, yfit)
```

**Example with Unsupervised Learning**

Choose model

```python
from sklearn.decomposition import PCA
```

Choose model hyperparameters

```python
model = PCA(n_components=2)
model
```

Create sample data

```python
import seaborn as sns
iris = sns.load_dataset('iris')
X_iris = iris.drop('species', axis = 1)
X_iris.shape
```

```python
y_iris = iris['species']
y_iris.shape
```

Apply the model on the data

```python
model.fit(X_iris)
```

Transform the data to two-dimensional

```python
X_2D = model.transform(X_iris)
```

Plot the data

```python
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
sns.lmplot(x="PCA1", y="PCA2", hue='species', data=iris, fit_reg=False)
```
